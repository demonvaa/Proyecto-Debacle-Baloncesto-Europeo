```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import unicodedata
import numpy as np
import unicodedata


df_campeones = pd.read_csv("Campeones_Copa_Europa_Euroleague.csv")
df_ico_modernos = pd.read_csv("Codigos_ICO_Modernos.csv")
df_ico_antiguos= pd.read_csv("Codigos_ICO_Antiguos.csv")
```

```python
columnas_originales = df_campeones.columns
columnas_finales = ["year_win","team", "statenme"]
diccionario_cambio_campeones = dict(zip(columnas_originales,columnas_finales))
df_campeones.rename(columns=diccionario_cambio_campeones, inplace=True)
df_campeones.head()
```

Le cambio columnas a esta tabla para unirlas luego en ingles

```python
columnas_originales = df_ico_antiguos.columns
columnas_finales = ["stateabb","statenme", "styear", "endyear"]
diccionario_cambio_campeones = dict(zip(columnas_originales,columnas_finales))
df_ico_antiguos.rename(columns=diccionario_cambio_campeones, inplace=True)
df_ico_antiguos.head()
```

```python
df_union_icos = pd.concat([df_ico_antiguos, df_ico_modernos], ignore_index=True, sort=False)
df_union_icos.head(1000
                   )
```

```python
df_union_icos["styear"] = pd.to_numeric(df_union_icos["styear"], errors="coerce").astype("Int64")
df_union_icos["endyear"] = pd.to_numeric(df_union_icos["endyear"], errors="coerce").astype("Int64")

cols_texto = df_union_icos.select_dtypes(include="object").columns
df_union_icos[cols_texto] = df_union_icos[cols_texto].fillna("")

```

```python
df_union_icos= df_union_icos.sort_values("statenme")
print(df_union_icos.shape)
df_union_icos.head()
 
```

Ahora aplico una columna a la tabla campeones donde añado el codigo y la nota segun el año

```python
lista_paises_campeones= df_campeones["statenme"].unique().tolist() 
for x in lista_paises_campeones:
    print(f"{x}: {x in df_union_icos['statenme'].values}")

```

Vamos por fin a unirlo con df_campeones

```python
df_campeones.head()

```

```python
df_tabla_final = df_campeones.merge(
    df_union_icos,
    on="statenme",
    how="left"
)
df_tabla_final.head(20)
```

```python
lista_original = df_tabla_final.columns.tolist()
lista_nueva = ['year_win','team','statenme']
df_tabla_final = df_tabla_final.rename(columns=dict(zip(lista_original, lista_nueva)))
df_tabla_final.head()
```

```python
df_tabla_final.to_csv("Copa_Europa_Euroleague.csv", index=False, encoding="utf-8")
```
