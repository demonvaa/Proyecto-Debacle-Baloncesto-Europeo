## 🇷🇺 URSS — Apogeo absoluto (1958–1971)
**Periodo dominante:** Copa Europa hasta 1992  
**Equipos:** ASK Riga, CSKA Moscú, Dinamo Tbilisi  

**Razones del apogeo:**
- Sistema deportivo estatal extremadamente profesionalizado.  
- Reclutamiento obligatorio de los mejores jugadores.  
- Entrenadores pioneros en defensa y preparación física.  
- Plantillas estables durante años.

**Periodo dentro de los 3 bloques:**
- **Hasta 1992 → APOGEO TOTAL**  
- 1992–2001 → No existe  
- Euroleague → No existe

## 🇾🇺 Yugoslavia — Apogeo tardío pero explosivo (1978–1991)
**Periodo dominante:** Copa Europa hasta 1992  
**Equipos:** Bosna, Cibona, Split (Jugoplastika), Partizan  

**Razones del apogeo:**
- Escuela yugoslava: técnica, creatividad, lectura del juego.  
- Cantera legendaria (Petrović, Kukoč, Rađa, Divac…).  
- Entrenadores de élite.  
- Competitividad interna altísima.

**Periodo dentro de los 3 bloques:**
- **Hasta 1992 → APOGEO TOTAL**  
- 1992–2001 → No existe (disolución)  
- Euroleague → No existe como país

## 🇮🇹 Italia — Apogeo en los 70 y 80 (1969–1988)
**Periodo dominante:** Copa Europa hasta 1992  
**Equipos:** Varese, Cantù, Milano  

**Razones del apogeo:**
- Inversión privada fuerte.  
- Fichajes americanos antes que otros países.  
- Liga italiana: la más fuerte de Europa en los 80.  
- Entrenadores tácticos y defensivos.

**Periodo dentro de los 3 bloques:**
- **Hasta 1992 → APOGEO TOTAL**  
- 1992–2001 → Declive  
- Euroleague → Residual

## 🇮🇱 Israel — Apogeo largo y estable (1977–2005)

**Equipo:** Maccabi Tel Aviv  

**Razones del apogeo:**
- Proyecto de club estable durante décadas.
